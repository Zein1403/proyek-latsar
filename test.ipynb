```python
import csv
import ipywidgets as widgets
from IPython.display import display, clear_output

# File name
filename = 'manual_data.csv'

# Options
satuan_options = ['Meter', 'kg', 'liter', 'buah']
tempat_options = ['Gudang lantai 1', 'Gudang A lantai 4 ', 'Gudang B lantai 4', 'Gudang lantai 6']

# Write header only if file doesn't exist
import os
if not os.path.exists(filename):
    with open(filename, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Nama', 'Jumlah', 'Satuan', 'Tempat'])

# Widgets for input
nama_input = widgets.Text(description="Nama:")
jumlah_input = widgets.BoundedIntText(description="Jumlah:", min=0, max=100000)
satuan_dropdown = widgets.Dropdown(options=satuan_options, description="Satuan:")
tempat_dropdown = widgets.Dropdown(options=tempat_options, description="Tempat:")

submit_button = widgets.Button(description="Submit", button_style='success')
update_button = widgets.Button(description="Update Item", button_style='warning')
display_button = widgets.Button(description="Show CSV", button_style='info')

output = widgets.Output()

# Function: Add item
def on_submit(b):
    with output:
        clear_output()
        nama = nama_input.value.strip()
        jumlah = jumlah_input.value
        satuan = satuan_dropdown.value
        tempat = tempat_dropdown.value

        if not nama:
            print("⚠️ Nama tidak boleh kosong.")
            return

        with open(filename, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow([nama, jumlah, satuan, tempat])
            print(f"✅ Data '{nama}' berhasil disimpan!")

        # Reset inputs
        nama_input.value = ''
        jumlah_input.value = 1

# Function: Update item
def on_update(b):
    with output:
        clear_output()
        target_name = nama_input.value.strip().lower()
        move_amount = jumlah_input.value
        satuan_value = satuan_dropdown.value
        new_location = tempat_dropdown.value

        if not target_name:
            print("⚠️ Masukkan Nama barang.")
            return

        updated_rows = []
        found = False
        merged_into_existing = False

        with open(filename, 'r', newline='') as f:
            reader = csv.DictReader(f)
            for row in reader:
                nama = row['Nama'].strip()
                satuan = row['Satuan'].strip()
                tempat = row['Tempat'].strip()
                jumlah = int(row['Jumlah'])

                is_target_item = nama.lower() == target_name and satuan == satuan_value

                # Case 1: Source match (remove from here)
                if is_target_item and tempat != new_location:
                    if jumlah >= move_amount:
                        found = True
                        remaining = jumlah - move_amount
                        if remaining >= 0:
                            row['Jumlah'] = str(remaining)
                            updated_rows.append(row)
                        # else: don't append — fully moved, so delete it
                        print(f"🔁 {move_amount} {satuan} dipindahkan dari {tempat} ke {new_location}")
                    else:
                        print(f"❌ Jumlah tidak cukup di {tempat}.")
                        updated_rows.append(row)

                # Case 2: Destination match (merge quantity)
                elif is_target_item and tempat == new_location:
                    row['Jumlah'] = str(jumlah + move_amount)
                    updated_rows.append(row)
                    merged_into_existing = True
                    found = True

                # Other rows stay the same
                else:
                    updated_rows.append(row)

        # If target row doesn't exist yet, add new one
        if found and not merged_into_existing:
            updated_rows.append({
                'Nama': nama_input.value.strip(),
                'Jumlah': str(move_amount),
                'Satuan': satuan_value,
                'Tempat': new_location
            })

        if found:
            with open(filename, 'w', newline='') as f:
                fieldnames = ['Nama', 'Jumlah', 'Satuan', 'Tempat','tanggal']
                writer = csv.DictWriter(f, fieldnames=fieldnames)
                writer.writeheader()
                writer.writerows(updated_rows)
            print("✅ Data berhasil diperbarui.")
        else:
            print("❌ Barang tidak ditemukan atau tidak dapat dipindahkan.")

# Function: Display CSV
def on_display(b):
    with output:
        clear_output()
        print("📄 Isi file CSV:")
        with open(filename, 'r') as f:
            content = f.read()
            print(content)

# Link functions
submit_button.on_click(on_submit)
update_button.on_click(on_update)
display_button.on_click(on_display)

# Layout
form = widgets.VBox([
    nama_input,
    jumlah_input,
    satuan_dropdown,
    tempat_dropdown,
    widgets.HBox([submit_button, update_button, display_button]),
    output
])

display(form)
```

```python
import csv
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime

# File name
filename = 'manual_data.csv'

# Options
satuan_options = ['Meter', 'kg', 'liter', 'buah']
tempat_options = ['Gudang lantai 1', 'Gudang A lantai 4 ', 'Gudang B lantai 4', 'Gudang lantai 6']

# Write header only if file doesn't exist
import os
if not os.path.exists(filename):
    with open(filename, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal'])

# Widgets for input
nama_input = widgets.Text(description="Nama:")
jumlah_input = widgets.BoundedIntText(description="Jumlah:", min=0, max=100000)
satuan_dropdown = widgets.Dropdown(options=satuan_options, description="Satuan:")
tempat_dropdown = widgets.Dropdown(options=tempat_options, description="Tempat:")

submit_button = widgets.Button(description="Submit", button_style='success')
update_button = widgets.Button(description="Update Item", button_style='warning')
display_button = widgets.Button(description="Show CSV", button_style='info')
decrease_button = widgets.Button(description="Decrease Item", button_style='danger')

output = widgets.Output()

# Function: Add item
def on_submit(b):
    with output:
        clear_output()
        nama = nama_input.value.strip()
        jumlah = jumlah_input.value
        satuan = satuan_dropdown.value
        tempat = tempat_dropdown.value

        if not nama:
            print("⚠️ Nama tidak boleh kosong.")
            return

        updated_rows = []
        found = False

        # Read existing rows and check if same item exists
        with open(filename, 'r', newline='') as f:
            reader = csv.DictReader(f)
            for row in reader:
                if (row['Nama'].strip().lower() == nama.lower() and
                    row['Satuan'].strip() == satuan and
                    row['Tempat'].strip() == tempat):
                    # Merge quantities
                    row['Jumlah'] = str(int(row['Jumlah']) + jumlah)
                    found = True
                updated_rows.append(row)

        if not found:
            # Add new row
            updated_rows.append({
                'Nama': nama,
                'Jumlah': str(jumlah),
                'Satuan': satuan,
                'Tempat': tempat
            })

        # Write updated rows
        with open(filename, 'a', newline='') as f:
             writer = csv.writer(f)
             tanggal = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
             writer.writerow([nama, jumlah, satuan, tempat, tanggal])

        print(f"✅ Data '{nama}' berhasil ditambahkan atau diperbarui.")

        # Reset inputs
        nama_input.value = ''
        jumlah_input.value = 1

def on_decrease(b):
    with output:
        clear_output()
        target_name = nama_input.value.strip()
        decrease_amount = jumlah_input.value
        satuan_value = satuan_dropdown.value
        location_value = tempat_dropdown.value

        if not target_name:
            print("⚠️ Masukkan nama barang yang ingin dikurangi.")
            return

        updated_rows = []
        item_found = False

        with open(filename, 'r', newline='') as f:
            reader = csv.DictReader(f)
            for row in reader:
                name_match = row['Nama'].strip().lower() == target_name.lower()
                unit_match = row['Satuan'].strip() == satuan_value
                location_match = row['Tempat'].strip() == location_value

                if name_match and unit_match and location_match:
                    current_qty = int(row['Jumlah'])
                    if current_qty < decrease_amount:
                        print(f"❌ Jumlah yang diminta ({decrease_amount}) melebihi stok ({current_qty}).")
                        return
                    elif current_qty == decrease_amount:
                        # Don't add this row back — item removed
                        print(f"🗑️ Item '{target_name}' dihapus dari lokasi karena jumlahnya jadi 0.")
                        item_found = True
                        continue
                    else:
                        row['Jumlah'] = str(current_qty - decrease_amount)
                        print(f"➖ Berhasil mengurangi {decrease_amount} {satuan_value} dari '{target_name}'.")
                        item_found = True
                updated_rows.append(row)

        if not item_found:
            print("❌ Item tidak ditemukan di lokasi dan satuan yang dipilih.")
            return

        with open(filename, 'w', newline='') as f:
            fieldnames = ['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal']
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(updated_rows)

# Function: Update item
def on_update(b):
    with output:
        clear_output()
        target_name = nama_input.value.strip()
        move_amount = jumlah_input.value
        satuan_value = satuan_dropdown.value
        new_location = tempat_dropdown.value

        if not target_name:
            print("⚠️ Masukkan Nama barang.")
            return

        target_name_lower = target_name.lower()
        source_total = 0
        updated_rows = []
        move_possible = False

        # Read all rows first
        with open(filename, 'r', newline='') as f:
            reader = list(csv.DictReader(f))

        # First pass: calculate total available in other locations
        for row in reader:
            if (row['Nama'].strip().lower() == target_name_lower and
                row['Satuan'].strip() == satuan_value and
                row['Tempat'].strip() != new_location):
                source_total += int(row['Jumlah'])

        if source_total < move_amount:
            print("❌ Jumlah total tidak mencukupi untuk dipindahkan.")
            return

        amount_to_move = move_amount
        for row in reader:
            name = row['Nama'].strip()
            satuan = row['Satuan'].strip()
            tempat = row['Tempat'].strip()
            jumlah = int(row['Jumlah'])

            # Source: match name + unit + NOT destination
            if (name.lower() == target_name_lower and satuan == satuan_value and tempat != new_location and amount_to_move > 0):
                if jumlah <= amount_to_move:
                    amount_to_move -= jumlah
                    # Don't add row (fully consumed)
                    continue
                else:
                    row['Jumlah'] = str(jumlah - amount_to_move)
                    amount_to_move = 0
                    updated_rows.append(row)
                    continue

            updated_rows.append(row)

        # Second pass: try to find if destination already has the item
        merged = False
        for row in updated_rows:
            if (row['Nama'].strip().lower() == target_name_lower and
                row['Satuan'].strip() == satuan_value and
                row['Tempat'].strip() == new_location):
                row['Jumlah'] = str(int(row['Jumlah']) + move_amount)
                merged = True
                break

        if not merged:
            # Add new row for destination
            
            updated_rows.append({
                'Nama': row['Nama'],
                'Jumlah': str(move_amount),
                'Satuan': satuan_value,
                'Tempat': new_location,
                'Tanggal': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            })

        with open(filename, 'w', newline='') as f:
            fieldnames = ['Nama', 'Jumlah', 'Satuan', 'Tempat','Tanggal']
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(updated_rows)

        print(f"✅ Berhasil memindahkan {move_amount} {satuan_value} ke {new_location}.")

# Function: Display CSV
def on_display(b):
    with output:
        clear_output()
        print("📄 Isi file CSV:")
        with open(filename, 'r') as f:
            content = f.read()
            print(content)

# Link functions
submit_button.on_click(on_submit)
update_button.on_click(on_update)
display_button.on_click(on_display)
decrease_button.on_click(on_decrease)
# Layout
form = widgets.VBox([
    nama_input,
    jumlah_input,
    satuan_dropdown,
    tempat_dropdown,
    widgets.HBox([submit_button, update_button, decrease_button, display_button]),
    output
])

display(form)
```

```python
from datetime import datetime
import os
import csv

# Define the full Streamlit app content as a Python script string
streamlit_app_code = ""
import streamlit as st
import csv
import os
from datetime import datetime

filename = 'inventory.csv'
log_filename = 'log_data.csv'

# Ensure the inventory CSV file exists
if not os.path.exists(filename):
    with open(filename, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal'])

# Ensure the log CSV file exists
if not os.path.exists(log_filename):
    with open(log_filename, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Tanggal', 'Aksi', 'Nama', 'Jumlah', 'Satuan', 'Dari', 'Ke'])

# Options
satuan_options = ['Meter', 'kg', 'liter', 'buah']
tempat_options = ['Gudang lantai 1', 'Gudang A lantai 4', 'Gudang B lantai 4', 'Gudang lantai 6']

# Streamlit UI
st.title("📦 Inventory Management")

st.subheader("Form Input")
nama = st.text_input("Nama")
jumlah = st.number_input("Jumlah", min_value=1, max_value=100000, step=1, value=1)
satuan = st.selectbox("Satuan", satuan_options)
tempat = st.selectbox("Tempat", tempat_options)
source = st.selectbox("Dari", tempat_options, key="source")
destination = st.selectbox("Ke", tempat_options, key="destination")

def log_action(action, nama, jumlah, satuan, source, destination):
    with open(log_filename, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([datetime.now().strftime('%Y-%m-%d %H:%M:%S'), action, nama, jumlah, satuan, source, destination])

# Submit
if st.button("Submit"):
    if not nama.strip():
        st.warning("⚠️ Nama tidak boleh kosong.")
    else:
        updated_rows = []
        found = False

        with open(filename, 'r', newline='') as f:
            reader = csv.DictReader(f)
            for row in reader:
                if (row['Nama'].strip().lower() == nama.lower() and row['Satuan'] == satuan and row['Tempat'] == tempat):
                    row['Jumlah'] = str(int(row['Jumlah']) + jumlah)
                    row['Tanggal'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    found = True
                updated_rows.append(row)

        if not found:
            updated_rows.append({
                'Nama': nama,
                'Jumlah': str(jumlah),
                'Satuan': satuan,
                'Tempat': tempat,
                'Tanggal': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            })

        with open(filename, 'w', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal'])
            writer.writeheader()
            writer.writerows(updated_rows)

        log_action("Submit", nama, jumlah, satuan, "-", tempat)
        st.success(f"✅ Data '{nama}' berhasil ditambahkan atau diperbarui.")

# Decrease
if st.button("Decrease Item"):
    if not nama.strip():
        st.warning("⚠️ Masukkan nama barang.")
    else:
        updated_rows = []
        found = False

        with open(filename, 'r', newline='') as f:
            reader = csv.DictReader(f)
            for row in reader:
                name_match = row['Nama'].strip().lower() == nama.lower()
                unit_match = row['Satuan'] == satuan
                location_match = row['Tempat'] == tempat

                if name_match and unit_match and location_match:
                    current_qty = int(row['Jumlah'])
                    if current_qty < jumlah:
                        st.error(f"❌ Jumlah ({jumlah}) melebihi stok ({current_qty}).")
                        return
                    elif current_qty == jumlah:
                        found = True
                        continue
                    else:
                        row['Jumlah'] = str(current_qty - jumlah)
                        row['Tanggal'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                        found = True
                updated_rows.append(row)

        if not found:
            st.error("❌ Item tidak ditemukan.")
            return

        with open(filename, 'w', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal'])
            writer.writeheader()
            writer.writerows(updated_rows)

        log_action("Decrease", nama, jumlah, satuan, tempat, "-")
        st.success(f"➖ Berhasil mengurangi {jumlah} {satuan} dari '{nama}' di {tempat}.")

# Update
if st.button("Update Item"):
    if source == destination:
        st.warning("⚠️ Lokasi asal dan tujuan tidak boleh sama.")
    elif not nama.strip():
        st.warning("⚠️ Masukkan nama barang.")
    else:
        updated_rows = []
        item_found = False

        with open(filename, 'r', newline='') as f:
            reader = list(csv.DictReader(f))

        for row in reader:
            if (row['Nama'].strip().lower() == nama.lower() and row['Satuan'] == satuan and row['Tempat'] == source):
                current_qty = int(row['Jumlah'])
                if current_qty < jumlah:
                    st.error(f"❌ Jumlah di {source} tidak cukup ({current_qty}).")
                    return
                elif current_qty == jumlah:
                    item_found = True
                    continue
                else:
                    row['Jumlah'] = str(current_qty - jumlah)
                    row['Tanggal'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    updated_rows.append(row)
                    item_found = True
                    continue
            updated_rows.append(row)

        if not item_found:
            st.error(f"❌ Item tidak ditemukan di {source}.")
            return

        # Add to destination
        merged = False
        for row in updated_rows:
            if (row['Nama'].strip().lower() == nama.lower() and row['Satuan'] == satuan and row['Tempat'] == destination):
                row['Jumlah'] = str(int(row['Jumlah']) + jumlah)
                row['Tanggal'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                merged = True
                break

        if not merged:
            updated_rows.append({
                'Nama': nama,
                'Jumlah': str(jumlah),
                'Satuan': satuan,
                'Tempat': destination,
                'Tanggal': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            })

        with open(filename, 'w', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=['Nama', 'Jumlah', 'Satuan', 'Tempat', 'Tanggal'])
            writer.writeheader()
            writer.writerows(updated_rows)

        log_action("Update", nama, jumlah, satuan, source, destination)
        st.success(f"✅ Berhasil memindahkan {jumlah} {satuan} dari {source} ke {destination}.")

# Display CSV
if st.button("Show Inventory"):
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            reader = csv.reader(f)
            data = list(reader)
            st.write("📄 Isi File Inventory:")
            st.dataframe(data)
    else:
        st.warning("⚠️ File CSV belum tersedia.")

if st.button("Show Log"):
    if os.path.exists(log_filename):
        with open(log_filename, 'r') as f:
            reader = csv.reader(f)
            data = list(reader)
            st.write("📑 Riwayat Aksi:")
            st.dataframe(data)
    else:
        st.warning("⚠️ File log belum tersedia.")


# Save to file
file_path = "/mnt/data/inventory_app.py"
with open(file_path, "w") as f:
    f.write(streamlit_app_code)

file_path  # Return the path to the user for download or further use

```
